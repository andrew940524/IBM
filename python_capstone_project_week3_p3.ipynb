```python
columns = ['PostalCode', 'Borough', 'Neighborhood']
df_neighbor = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df_neighbor = df_neighbor[0]
df_neighbor.head()
```

```python
df_neighbor = df_neighbor[df_neighbor.Borough != 'Not assigned']
df_neighbor.groupby('Postal code')
df_neighbor.reset_index(drop=True, inplace=True)
df_neighbor.replace(regex=['/'], value=",", inplace = True)
df_neighbor.head(12)
```

```python
df_neighbor.shape
```

```python
df_geospatial = pd.read_csv("http://cocl.us/Geospatial_data")
df_toronto_coord = df_neighbor.merge(df_geospatial, left_on='Postal code', right_on='Postal Code')
df_toronto_coord.drop(['Postal Code'], axis=1, inplace = True)
df_toronto_coord
```

### Plot the points on a map

```python
df_toronto_coord['location'] = list(zip(df_toronto_coord.Latitude, df_toronto_coord.Longitude))
df_toronto_coord
```

```python
import folium
torMap = folium.Map(location=[43.6487,-79.38544], zoom_start=12)
torMap
```

```python
neighbors = folium.map.FeatureGroup()

#loop through the neighborhoods
for lat, lng, in zip(df_toronto_coord.Latitude, df_toronto_coord.Longitude):
    neighbors.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

torMap.add_child(neighbors)
```
